### Deteksi Ukuran Gambar (`gundul.jpg`)

Bagian ini akan menampilkan dimensi gambar dalam piksel dan, karena skala telah ditentukan, juga dalam satuan dunia nyata.

```
import cv2
import numpy as np
from PIL import Image

image_path = '/content/gundul.jpg'

# Load the image using PIL first to handle various image types, then convert to numpy array
# It's better to use PIL for initial loading if the file type might be problematic for cv2.imread directly
img_pil = Image.open(image_path)
img = np.array(img_pil)

print("Melakukan deteksi ukuran gambar...")

# Mendapatkan dimensi gambar (tinggi, lebar, channels)
height, width = img.shape[:2]

print(f"Dimensi gambar dalam piksel: Lebar = {width} piksel, Tinggi = {height} piksel")

# Jika scale_factor_per_pixel telah ditentukan sebelumnya, hitung ukuran dunia nyata
if 'scale_factor_per_pixel' in locals() and 'real_world_unit' in locals():
    real_world_width = width * scale_factor_per_pixel
    real_world_height = height * scale_factor_per_pixel

    print(f"Dimensi gambar dalam dunia nyata (menggunakan skala 1 piksel = {scale_factor_per_pixel:.4f} {real_world_unit}):")
    print(f"  Lebar = {real_world_width:.2f} {real_world_unit}")
    print(f"  Tinggi = {real_world_height:.2f} {real_world_unit}")

    # Hitung total luas gambar dalam dunia nyata
    total_real_world_area = real_world_width * real_world_height
    print(f"  Total Luas Gambar = {total_real_world_area:.2f} {real_world_unit}^2")
else:
    print("Faktor skala belum ditentukan, sehingga tidak dapat menghitung dimensi dalam unit dunia nyata.")
    print("Silakan jalankan sel untuk 'Menentukan Skala dari Gambar' terlebih dahulu.")
```
